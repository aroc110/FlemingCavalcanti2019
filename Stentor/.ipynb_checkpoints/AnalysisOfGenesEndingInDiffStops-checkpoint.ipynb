# Calculate Tandem Stop Codons Overepresentation for Genes ending in different stop codons

```python
import os
import pandas as pd
import math
utrSize = 20
numberOfPositions = utrSize // 3

filein = open('Stentor_coeruleus.utr20','r')
stops = ['TAA','TAG','TGA']
#Calculate frequency of stops in the 3' UTRs

tandems = []

filein.readline()
for line in filein:
    num, name, realStop, utr, seq = line.split()
    utr = utr.upper()
    tandem = 0
    for i in range(numberOfPositions):
        current = utr[i*3:i*3+3]
        if current in stops:
            tandem = i + 1
            break
    found = 0
    if tandem:
        found = 1
    tandems.append([name, realStop,tandem,found,1])

filein.close()

tandemDF = pd.DataFrame(tandems, columns=['gene','stop','tandemPos',
                                         'hasTandem','counter'])
```

```python
tandemDF
```

```python
sumTandem = tandemDF.groupby('stop').sum()
```

```python
sumTandem['ratio']=sumTandem['hasTandem']/sumTandem['counter']
```

```python
sumTandem.to_csv('tandemTable.tsv',sep='\t')
sumTandem
```

```python
obsTSC = sumTandem['hasTandem']['TGA']
obsNo = sumTandem['counter']['TGA'] - sumTandem['hasTandem']['TGA']
expFreq = ((sumTandem['hasTandem']['TAA']+sumTandem['hasTandem']['TAG'])/
          (sumTandem['counter']['TAA']+sumTandem['counter']['TAG']))
expTSC = sumTandem['counter']['TGA'] * expFreq
expNo = sumTandem['counter']['TGA'] * (1-expFreq)

print(obsTSC, obsNo)
print(expTSC, expNo)
```

```python
from scipy.stats import chisquare
obs = [obsTSC,obsNo]
exp = [expTSC,expNo]
chisquare(obs,exp)
```
