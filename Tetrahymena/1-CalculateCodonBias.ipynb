# Calculate the values for CAI for each codon using high frequency genes

For each amino acid in a gene, the weight of each of its codons represented by a parameter termed relative adaptiveness (wi), is computed from a reference sequence set, as the ratio between the observed frequency of the codon fi and the frequency of the most frequent synonymous codon fj for that amino acid.

$$ {\displaystyle w_{i}={\frac {f_{i}}{\max(f_{j})}}\qquad i,j\in [{\text{synonymous codons for amino acid}}]} $$

The CAI of a gene is simply defined as the geometric mean of the weight associated to each codon over the length (L) of the gene sequence (measured in codons).[2]

$$ {\displaystyle {\text{CAI}}=(\Pi _{i=1}^{L}w_{i})^{\frac {1}{L}}} $$

```python
import pandas as pd
from Bio import SeqIO
import re
import functionsUTR as utr
fasta_dic = utr.parse_fasta('T_thermophila_June2014_CDS.fasta')
```

```python
filein = open("referenceSetForCAI.txt")
caiGenes = []
count1 = 0
count2 = 0
for line in filein:
    name = line.strip()
    count1 += 1
    if name in fasta_dic:
        caiGenes.append(name)
        count2 += 1
    else:
        print(name)
print(count1,count2)
```

```python
codons = {}
for gene in caiGenes:
    seq = fasta_dic[gene]
    if len(seq) % 3 == 0:
        for i in range(len(seq) // 3):
            codon = seq[i*3:i*3 + 3]
            codons[codon] = codons.get(codon,0) + 1
    else:
        print('problem')
len(codons)
```

```python
syncodons = { #for tetrahymena - only one stop codon: TGA
   'CYS': ['TGT', 'TGC'], 
   'ASP': ['GAT', 'GAC'], 
   'SER': ['TCT', 'TCG', 'TCA', 'TCC', 'AGC', 'AGT'], 
   'GLN': ['CAA', 'CAG', 'TAG', 'TAA'], 
   'MET': ['ATG'], 
   'ASN': ['AAC', 'AAT'], 
   'PRO': ['CCT', 'CCG', 'CCA', 'CCC'], 
   'LYS': ['AAG', 'AAA'], 
   'STOP': ['TGA'], 
   'THR': ['ACC', 'ACA', 'ACG', 'ACT'], 
   'PHE': ['TTT', 'TTC'], 
   'ALA': ['GCA', 'GCC', 'GCG', 'GCT'], 
   'GLY': ['GGT', 'GGG', 'GGA', 'GGC'], 
   'ILE': ['ATC', 'ATA', 'ATT'], 
   'LEU': ['TTA', 'TTG', 'CTC', 'CTT', 'CTG', 'CTA'], 
   'HIS': ['CAT', 'CAC'], 
   'ARG': ['CGA', 'CGC', 'CGG', 'CGT', 'AGG', 'AGA'], 
   'TRP': ['TGG'], 
   'VAL': ['GTA', 'GTC', 'GTG', 'GTT'], 
   'GLU': ['GAG', 'GAA'], 
   'TYR': ['TAT', 'TAC'] } 
```

```python
wi = {}
for aa, syn in syncodons.items():
    total = 0
    synon = []
    
    for i in syn:
        synon.append(codons[i])
        total += codons[i]
    denominator = max(synon)
    
    for i in syn:
        wi[i] = codons[i]/denominator
        
    check = 0
    for i in syn:
        check += wi[i]
    print(aa, check, total)
```

## Calculate CAI for each Tetrahymena gene

CAI formula is:

$$ {\displaystyle {\text{CAI}}=(\Pi _{i=1}^{L}w_{i})^{\frac {1}{L}}} $$

I'm calculating as:


$$ \ln { (CAI) } =\ln { ((\Pi _{ i=1 }^{ L }w_{ i })^{ \frac { 1 }{ L }  } } )=\frac { \ln { ((\Pi _{ i=1 }^{ L }w_{ i }) }  }{ L } =\frac { \sum _{ i=1 }^{ L }{ \ln { ({ w }_{ i }) }  }  }{ L } $$

Taking the exponantial on both sides gives:

$$ CAI={ e }^{ \frac { \sum _{ i=1 }^{ L }{ \ln { ({ w }_{ i }) }  }  }{ L }  } $$

```python
import os
import pandas as pd
from Bio import SeqIO
import re
import functionsUTR as utr
import math

all_genes = utr.parse_fasta('T_thermophila_June2014_CDS.fasta')
```

```python
allCAI = []
countAll = 0
for name, sequence in all_genes.items():
    countAll += 1
    CAI = 1
    realLength = 0
    for i in range(len(sequence) // 3):
        codon = sequence[i*3:i*3+3]
        if codon in wi:
            realLength += 1
            CAI += lwi[codon]
    CAI = math.exp( CAI / realLength )
    allCAI.append([name,CAI])
    print(countAll, name, CAI, len(sequence)//3, realLength)

CAIs = pd.DataFrame(allCAI, columns = ['gene', 'CAI'])
```
